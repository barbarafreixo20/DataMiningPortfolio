```python
import pandas as pd
import numpy as np
from scipy.stats import f_oneway

def f_classif(dataset):
    # Substitui as features categóricas por one-hot encoding
    dataset = pd.get_dummies(dataset)

    # Separa as amostras por classe
    classes = dataset.iloc[:,-1].unique()
    groups = [dataset[dataset.iloc[:, -1] == c].iloc[:, :-1] for c in classes]

    # Calcula os valores de F e p
    f_values, p_values = f_oneway(*groups)

    return f_values, p_values
```

```python
# Carrega o arquivo CSV para um dataframe do pandas
dataset = pd.read_csv('titanic.csv')

# Chama a função f_classif para calcular os valores de F e p
f_values, p_values = f_classif(dataset)

# Imprime os valores de F e p
print(f'Valores de F: {f_values}')
print(f'Valores de p: {p_values}')
```
